```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo Excel (reemplaza por tu nombre de archivo)
archivo = "problem4_solucion.xlsx"
df = pd.read_excel(archivo)

# Asegúrate de que el archivo tiene al menos 3 columnas
# Columnas se indexan desde 0
x = df.iloc[:, 0]  # Columna 1
y = df.iloc[:, 2]  # Columna 3

# Graficar
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='-', color='blue')
plt.xlabel(df.columns[0])
plt.ylabel(df.columns[2])
plt.title("Gráfico: Columna 1 vs. Columna 3")
plt.grid(True)
plt.tight_layout()
plt.show()
```
