## Mezcla de aguas

En geoquímica acuosa, existen muchas situaciones en las que diferentes tipos de agua se mezclan. En estas prácticas estudiaremos ejemplos como la **intrusión marina en acuíferos costeros** y la **mezcla de aguas superficiales** (ríos, lagos, etc.).

El proceso de mezcla entre dos aguas distintas genera un **desequilibrio** en la nueva solución, lo cual puede activar procesos geoquímicos modificadores, como la **precipitación o disolución de calcita**, **reducción de sulfatos**, entre otros. La variación en las concentraciones después de la mezcla (también llamada **delta iónica**) es clave para evaluar si un elemento se comporta de manera conservativa durante la mezcla y detectar posibles procesos modificadores.

---

### En esta práctica abordaremos los siguientes problemas geoquímicos:

- Desarrollo de **sistemas kársticos costeros** en acuíferos de caliza por intrusión marina.
- **Reacciones de disolución–precipitación** y **neutralización** en el agua resultante.
- **Cuantificación de la mezcla** de aguas de manera conservativa.
- **Determinación de proporciones de mezcla** utilizando elementos conservativos como el **Cl⁻**.

---

Todas estas combinaciones se construyen a partir de una plantilla similar a la usada en la sección *Especiación de un agua natural*, pero utilizando la palabra clave `MIX`. El archivo de entrada está compuesto por tres partes:

1. Características del **agua dulce**.
2. Características del **agua salada (marina)**.
3. Definición de las **proporciones de mezcla**.

---

```phreeqc

````python
import subprocess

# Step 1: Create the PHREEQC input file
input_file_content = """
TITLE Mezcla de aguas (Mixing).
SOLUTION 1  Agua dulce
    temp      25
    pH        7.10
    units     mg/L
    density   1.00
    Ca        80
    Mg        7.1
    Na        14
    K         3.3
    Alkalinity 253. as HCO3
    S(6)      1.8
    Cl        34
SAVE solution 1
END
SOLUTION 2  agua de mar
    temp    25.0
     pH      8.22
    units    ppm
    density   1.023
    Ca       412.3
    Mg       1291.8
    Na       10768.0
    K         399.1
    Cl        19353.0
    Alkalinity      141.682 as HCO3
    S(6)            2712.0
SAVE solution 2
END
TITLE Mezcla de las aguas: definición de las proporciones.
MIX 1
    1      0.9
    2      0.1
SAVE solution   3
END
"""

# Save the input file
input_file_name = "phreeqc_example10.pqi"
with open(input_file_name, "w") as file:
    file.write(input_file_content)
print(f"PHREEQC input file '{input_file_name}' created successfully.")

# Step 2: Run PHREEQC using subprocess
output_file_name = "phreeqc_example10_out.txt"
database_file = "/opt/shared/PHREEQC_code/phreeqc-3.8.6-17100/database/phreeqc.dat"  # Update the path if necessary
phreeqc_executable = "/opt/shared/PHREEQC_code/phreeqc-3.8.6-17100/src/phreeqc"  # Use "phreeqc.exe" on Windows, or the full path to the executable

# Run PHREEQC
try:
    subprocess.run([phreeqc_executable, input_file_name, output_file_name, database_file], check=True)
    print(f"PHREEQC run completed. Output saved in '{output_file_name}'.")
except subprocess.CalledProcessError as e:
    print(f"PHREEQC execution failed: {e}")    

# Display the contents of the output file, ignoring problematic characters
try:
    with open(output_file_name, "r", encoding="utf-8", errors="ignore") as output_file:
        output_content = output_file.read()
    print("PHREEQC Output:\n")
    print(output_content)
except FileNotFoundError:
    print(f"Output file '{output_file_name}' not found.")
````

Debido a la relación no lineal (cúbica) entre la solubilidad de la calcita, la salinidad, la pCO₂, el pH y la temperatura, la mezcla de dos aguas con diferentes estados de saturación puede dar lugar a una **agua mezclada que esté sobresaturada o subsaturada**, dependiendo de estas características fisicoquímicas y de la proporción de mezcla.

Ejemplos comunes incluyen el desarrollo de **sistemas kársticos en acuíferos de calcita** dentro de la zona de mezcla entre agua de mar y agua del acuífero (ambas saturadas con respecto a la calcita).
