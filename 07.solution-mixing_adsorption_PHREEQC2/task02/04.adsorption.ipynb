## Intercambio catiónico

Las reacciones de intercambio iónico se describen considerando los **cationes intercambiables** o adsorbatos y el **adsorbente** o intercambiador (superficie del mineral). En el intercambio iónico, los iones están **débilmente adsorbidos** (iones lábiles), de modo que pueden ser reemplazados **reversiblemente** por otros iones, conservando la carga. Este proceso es de **cinética rápida**, como se abordará en el tema de *Cinética de los procesos geoquímicos*.

Muchos minerales con alta capacidad de intercambio, a los valores de pH típicos de las aguas naturales, presentan **superficies cargadas negativamente**, como las **arcillas** o los **óxidos de manganeso**, que tienen un **pHPZC bajo**. En estos casos, el intercambio ocurre entre cationes, por lo que el proceso se denomina **intercambio catiónico**. Por ejemplo, el intercambio entre **Na⁺** y **K⁺** (intercambio homovalente) en la superficie de una montmorillonita (denominada como X) puede representarse como:

$$
\mathrm{Na^+ + K{-}X \rightleftharpoons Na{-}X + K^+}
$$

Las reacciones de intercambio catiónico se basan en el **equilibrio termodinámico** entre los iones en disolución y en la fase adsorbida, y pueden expresarse como una constante de equilibrio:

$$
K_{Na/K} = \frac{a_{K} \cdot a_{Na-X}}{a_{Na} \cdot a_{K-X}}
$$

donde $(a_{K}$) y $(a_{Na}$) son las **actividades** de K y Na en disolución. Los coeficientes de actividad se pueden obtener usando las teorías implementadas en PHREEQC, como la teoría de asociación iónica, interacción específica o el modelo de Pitzer.

No existe un único criterio para definir la **actividad del catión adsorbido** ($(a_{Na-X}, a_{K-X}$)); pueden usarse fracciones molares o equivalentes. El cálculo se complica cuando hay **intercambios heterovalentes** (por ejemplo, entre Ca²⁺ y Na⁺). Hay tres modelos principales:

- **Gaines-Thomas** (por defecto en PHREEQC, usa fracciones equivalentes)
- **Vanselow** (usa fracciones molares)
- **Gapon** (común en geoquímica de suelos, no recomendable para intercambios heterovalentes)

Para más detalles, consultar *Appelo y Postma (2005) section 6.3*.

### PHREEQC y el intercambio catiónico

PHREEQC calcula la **especiación** y **concentraciones** de las especies intercambiadas tanto en disolución como en el adsorbente (por ejemplo, Ca-X, Na-X). Estos cálculos pueden realizarse manteniendo fija la composición acuosa o permitiendo su ajuste.

PHREEQC utiliza tres palabras clave principales:

- `EXCHANGE`: define cantidad y composición del intercambiador
- `EXCHANGE_MASTER_SPECIES`: define especies maestras de intercambio
- `EXCHANGE_SPECIES`: define estequiometrías y constantes log K

La composición inicial del intercambio puede definirse:

1. **Explícitamente**, listando las especies adsorbidas y disueltas.
2. **Implícitamente**, estableciendo el equilibrio con una solución acuosa dada.

El número de sitios de intercambio puede:

- Ser fijo,
- Depender de una fase de equilibrio,
- O depender de un reactivo cinético.

Los datos de intercambio están incluidos en las bases `phreeqc.dat` y `wateq4f.dat`, y pueden modificarse en el archivo de entrada.

PHREEQC permite trabajar con múltiples intercambiadores en equilibrio y manejar cambios en el número de sitios si hay precipitación/disolución de minerales. El usuario puede decidir si la variación es proporcional a la fase mineral o a un reactivo cinético.

---

### Cálculo de la cantidad de intercambiador (X)

En un **material suelto** (como un suelo en un experimento batch):

$$
X~(\mathrm{moles}) = 10^{-5} \cdot \mathrm{CEC~(meq/100g)} \cdot m_{\text{sol}}~(\mathrm{g})
$$

Ejemplo: para 10 g de suelo con CEC = 20 meq/100 g  
$$
X = 10^{-5} \cdot 20 \cdot 10 = 0.002~\mathrm{moles}
$$

En un **material poroso**:

$$
X = 10^{-5} \cdot \mathrm{CEC} \cdot \left( \frac{\rho_b}{\theta} \right)
$$

donde:
- $( \rho_b $) = densidad del sólido,
- $( \theta $) = porosidad.

---

### Estimaciones del CEC

El CEC puede obtenerse experimentalmente mediante experimentos batch y se expresa por 100 g de sólido seco.

Una expresión empírica para suelos (Breeuwsma, 1986):

$$
\mathrm{CEC~(meq/100g)} = 0.7 \cdot \% \text{arcillas} + 3.5 \cdot \% \text{carbón orgánico}
$$

---

### Selectividad de los intercambiadores

- Generalmente, los intercambiadores muestran mayor selectividad por cationes **divalentes** que por **monovalentes**:  
  $$
  \text{Ca}^{2+} > \text{Na}^{+}
  $$
  
- Para cationes con igual carga, prefieren los de **mayor radio iónico (menor radio hidratado)**:  
  $$
  \text{Ba}^{2+} > \text{Ca}^{2+}, \quad \text{Cs}^{+} > \text{Na}^{+}
  $$

- La cantidad y dirección del intercambio también dependen de:
  - Concentración de iones en disolución
  - Fuerza iónica
  - pH
  - Temperatura
  - Propiedades del intercambiador: CEC, superficie específica, química superficial, etc.


### Ejemplo: Variación de la composición del agua por intercambio catiónico

En este ejemplo se cuantifica cómo **se modifica la composición de un agua** al entrar en equilibrio con un sólido que contiene **cationes previamente adsorbidos en su superficie**. El intercambiador `X` contiene inicialmente **Ca²⁺**, **Na⁺** y **K⁺**, con un total de **6.6·10⁻³ moles** de sitios de intercambio.

---

#### 📥 Entrada PHREEQC

```phreeqc
TITLE Variación de la composición del agua por intercambio catiónico
SOLUTION 1 
    temp      25
    units     mol/kgw
    Ca        0.0001
    K         0.0002
    Na        0.001
    Cl        0.0014 charge     # ajusta la carga con Cl

EXCHANGE 1  
    CaX2      3e-3
    NaX       5e-4
    KX        1e-4

PRINT
    -reset    false
    -totals   true
    -exchange true
END


````python
import subprocess

# Step 1: Create the PHREEQC input file
input_file_content = """
TITLE Variación de la composición del agua por intercambio catiónico
SOLUTION 1 
temp      25;   units     mol/kgw
Ca       0.0001; K         0.0002; Na        0.001
Cl        0.0014 charge      #ajusta la carga con el Cl
EXCHANGE 1  
   CaX2        3e-3
   NaX         5e-4
   KX          1e-4
END
"""

# Save the input file
input_file_name = "phreeqc_example10.pqi"
with open(input_file_name, "w") as file:
    file.write(input_file_content)
print(f"PHREEQC input file '{input_file_name}' created successfully.")

# Step 2: Run PHREEQC using subprocess
output_file_name = "phreeqc_example10_out.txt"
database_file = "/opt/shared/PHREEQC_code/phreeqc-3.8.6-17100/database/phreeqc.dat"  # Update the path if necessary
phreeqc_executable = "/opt/shared/PHREEQC_code/phreeqc-3.8.6-17100/src/phreeqc"  # Use "phreeqc.exe" on Windows, or the full path to the executable

# Run PHREEQC
try:
    subprocess.run([phreeqc_executable, input_file_name, output_file_name, database_file], check=True)
    print(f"PHREEQC run completed. Output saved in '{output_file_name}'.")
except subprocess.CalledProcessError as e:
    print(f"PHREEQC execution failed: {e}")    

# Display the contents of the output file, ignoring problematic characters
try:
    with open(output_file_name, "r", encoding="utf-8", errors="ignore") as output_file:
        output_content = output_file.read()
    print("PHREEQC Output:\n")
    print(output_content)
except FileNotFoundError:
    print(f"Output file '{output_file_name}' not found.")
````
